```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage import segmentation
import cv2

%matplotlib inline
```

```python
image = plt.imread('test.png')
```

```python
plt.imshow(image,cmap=plt.cm.gray)
image.shape
```

```python
seg = segmentation.felzenszwalb(image,scale=5,sigma=.25,min_size=50)
plt.imshow(seg)
print(np.max(seg))
```

```python
buf = 3
p = []
for i in range(441):
    x1,x2,y1,y2 = bbox2(seg==i)
    p.append(image[x1-buf:x2+buf,y1-buf:y2+buf])

from IPython.display import clear_output
```

```python
labels = []
for i in range(len(p)):
    plt.imshow(p[i],cmap=plt.cm.gray)
    plt.show()
    labels.append(input("What character is this? (Enter -1 for unrecognizable or double character)"))
    clear_output()
    if i == 10:
        break
```

```python
images = [p[i] for i in np.where(np.array(labels) != '-1')[0]]
for i in images:
    plt.imshow(i,cmap=plt.cm.gray)
    plt.show()
```

```python
def bbox2(img):
    rows = np.any(img, axis=1)
    cols = np.any(img, axis=0)
    rmin, rmax = np.where(rows)[0][[0, -1]]
    cmin, cmax = np.where(cols)[0][[0, -1]]

    return rmin, rmax, cmin, cmax

plt.figure(1,figsize=(10,10))
buf = 3
for i in range(441):
    x1,x2,y1,y2 = bbox2(seg==i)
    plt.subplot(21,21,i+1)
    plt.imshow(image[x1-buf:x2+buf,y1-buf:y2+buf],cmap=plt.cm.gray)
    plt.axis('off')
```
